```python
Q1. What are the three measures of central tendency?
The three measures of central tendency are:
1.	Mean: The average of all data points.
2.	Median: The middle value when the data is sorted.
3.	Mode: The most frequent value in the dataset.

    
    
    
    Q2. What is the difference between the mean, median, and mode? How are they used to measure the central tendency of a dataset?
•	Mean: The sum of all values divided by the number of values. It gives the overall average of the dataset but can be affected by extreme values (outliers).
o	Use: Best for symmetric datasets without outliers.
•	Median: The middle value when data is ordered. It is resistant to outliers and skewed data.
o	Use: Useful when data is skewed or contains outliers.
•	Mode: The most frequent value in the dataset. It is especially useful for categorical data or to identify the most common value in a dataset.
o	Use: Suitable when you want to know the most frequent observation.

    
    
Q3. Measure the three measures of central tendency for the given height data:
Data: [178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,176.5][178, 177, 176, 177, 178.2, 178, 175, 179, 180, 175, 178.9, 176.2, 177, 172.5, 178, 176.5][178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,176.5]

import numpy as np
data = [178, 177, 176, 177, 178.2, 178, 175, 179, 180, 175, 178.9, 176.2, 177, 172.5, 178, 176.5]

# Calculating mean, median, mode
mean = np.mean(data)
median = np.median(data)
from scipy import stats
mode = stats.mode(data)

mean, median, mode
•	Mean: 177.07
•	Median: 177
•	Mode: 178



Q4. Find the standard deviation for the given data:
Data: [178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,176.5][178, 177, 176, 177, 178.2, 178, 175, 179, 180, 175, 178.9, 176.2, 177, 172.5, 178, 176.5][178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,176.5]

# Standard Deviation
std_dev = np.std(data)
std_dev
•	Standard Deviation: 1.835



Q5. How are measures of dispersion such as range, variance, and standard deviation used to describe the spread of a dataset? Provide an example.
•	Range: The difference between the maximum and minimum values in a dataset. It provides a simple measure of spread.
o	Example: For the data [5,8,10,12,15][5, 8, 10, 12, 15][5,8,10,12,15], the range is 15−5=1015 - 5 = 1015−5=10.
•	Variance: The average of the squared differences from the mean. It measures how far data points spread out from the mean.
o	Example: For the data [5,8,10,12,15][5, 8, 10, 12, 15][5,8,10,12,15], variance is the average of the squared differences from the mean.
•	Standard Deviation: The square root of variance. It gives a sense of how much the data deviates from the mean in the same units as the data.
o	Example: If the variance is 4, then the standard deviation is 4=2\sqrt{4} = 24=2.



Q6. What is a Venn diagram?
A Venn diagram is a visual representation of the relationships between different sets. Each set is represented by a circle, and overlapping areas show elements that are common between sets.
Q7. For the two given sets A = {2,3,4,5,6,7} & B = {0,2,6,8,10}, find:
(i) A∩BA \cap BA∩B (Intersection): Common elements between sets A and B.
•	Answer: {2,6}\{2, 6\}{2,6}
(ii) A∪BA \cup BA∪B (Union): All elements from both sets, without repetition.
•	Answer: {0,2,3,4,5,6,7,8,10}\{0, 2, 3, 4, 5, 6, 7, 8, 10\}{0,2,3,4,5,6,7,8,10}



Q8. What do you understand about skewness in data?
Skewness refers to the asymmetry in the distribution of data. It can be:
•	Positive (Right-skewed): Tail is on the right side, with more data points concentrated on the left.
•	Negative (Left-skewed): Tail is on the left side, with more data points concentrated on the right.

    
Q9. If data is right-skewed, what will be the position of the median with respect to the mean?
In a right-skewed distribution, the mean is greater than the median because the long tail pulls the mean to the right.

    
Q10. Explain the difference between covariance and correlation. How are these measures used in statistical analysis?
•	Covariance: Measures how two variables vary together. It can take any value and is sensitive to the scale of the variables.
o	Example: If two variables increase together, they have a positive covariance.
•	Correlation: A standardized measure of the relationship between two variables, ranging from -1 to +1. It indicates the strength and direction of a linear relationship.
o	Example: A correlation of 0.8 suggests a strong positive linear relationship.

    

Q11. What is the formula for calculating the sample mean? Provide an example calculation for a dataset.
The sample mean is calculated as:
Sample Mean=∑xin\text{Sample Mean} = \frac{\sum{x_i}}{n}Sample Mean=n∑xi
Where xix_ixi are the data points and nnn is the number of observations.
•	Example: For the dataset [2,4,6,8][2, 4, 6, 8][2,4,6,8], the sample mean is:
                   2+4+6+84=5\frac{2 + 4 + 6 + 8}{4} = 542+4+6+8=5

    
Q12. For a normal distribution, what is the relationship between its measures of central tendency?
In a normal distribution, the mean, median, and mode are all equal and located at the center of the distribution.
Q13. How is covariance different from correlation?
Covariance measures how two variables vary together but is sensitive to the units of the variables, while correlation standardizes this relationship between -1 and 1, making it easier to interpret.

                                                                                                  
Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.
Outliers can significantly impact:
•	Mean: Outliers pull the mean toward them, making it less representative of the dataset.
o	Example: In the dataset [10,12,13,15,200][10, 12, 13, 15, 200][10,12,13,15,200], the mean is 50, which is not a good representation of the majority of the data points.
•	Median: The median is less affected by outliers since it only depends on the middle value.
•	Standard Deviation: Outliers increase the standard deviation, indicating higher spread in the data.

```
